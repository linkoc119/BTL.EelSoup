```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html

# Read the CSV file
df = pd.read_csv('data_scrape.csv')

# Create a Dash app
app = Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Scatter Plot of Views from Data"),
    dcc.Graph(
        id='scatter-plot',
        # Create the plotly scatter plot
        figure=px.scatter(
            df,
            y='Số lượt xem',  # Use 'views' for the x-axis
            x='Tên kênh',  # Use 'title' for the y-axis (or any other column you prefer)
            title="Scatter Plot of Views",
            labels={"Số lượt xem": "Number of Views", "Tên kênh": "Title"},
            hover_data={'Số lượt xem': True, 'Tiêu đề': True}
        )
    )
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
